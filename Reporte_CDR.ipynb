```python
def getNow():

	dt = datetime.now()

	yyn = str(dt.year) 

	if dt.month <= 9: mm = "0" + str(dt.month) 
	else: mm = str(dt.month) 

	if dt.day <= 9: dd = "0" + str(dt.day) 
	else: dd = str(dt.day) 

	if dt.hour <= 9: hh = "0" + str(dt.hour) 
	else: hh = str(dt.hour) 

	if dt.minute <= 9: mi = "0" + str(dt.minute) 
	else: mi = str(dt.minute) 

	if dt.second <= 9: ss = "0" + str(dt.second) 
	else: ss = str(dt.second) 

	now = yyn + mm + dd + hh + mi + ss

	return now
```
